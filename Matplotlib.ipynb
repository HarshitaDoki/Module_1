```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 45, 20, 10])
m=['apples','mangoes','banana','kiwi']
plt.title("famous fruits")
plt.pie(y,labels=m)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 45, 20, 10])
m=['apples','mangoes','banana','kiwi']
e=[0,0.1,0,0]
plt.title("famous fruits")
plt.pie(y,labels=m,explode=e)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 45, 20, 10])
m=['apples','mangoes','banana','kiwi']
e=[0,0.1,0,0]
c=['yellow','magenta','cyan','green']

plt.title("famous fruits")
plt.pie(y,labels=m,startangle=90,shadow= True,explode=e,colors=c)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 45, 20, 10])
m=['apples','mangoes','banana','kiwi']
plt.title("famous fruits")
plt.pie(y,labels=m)
plt.legend(title='Fruits')
plt.show() 
```
